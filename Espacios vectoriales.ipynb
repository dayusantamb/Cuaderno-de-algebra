# ***Espacios Vectoriales***
**Definición y ejemplos de espacios vectoriales**

Un espacio vectorial es una estructura algebraica que consiste en un conjunto de elementos llamados vectores, junto con dos operaciones: la suma vectorial y la multiplicación por un escalar. 

Es útil comprender esta definición a través de ejemplos concretos. Ejemplos comunes de espacios vectoriales:

1. Espacio R^n: El espacio R^n representa todas las n-tuplas de números reales. Por ejemplo, en R^2, los vectores serían parejas ordenadas de números reales (x, y), donde x e y pueden tomar cualquier valor real. La suma vectorial se realiza componente por componente, y la multiplicación por un escalar se aplica a cada componente individualmente.

2. Espacio de matrices: El espacio de matrices, como el conjunto de todas las matrices de tamaño m x n, forma un espacio vectorial. La suma de matrices se realiza sumando las entradas correspondientes, y la multiplicación por un escalar implica multiplicar cada entrada por ese escalar.

3. Espacio de polinomios: El conjunto de todos los polinomios de grado n o menor forma un espacio vectorial. 

4. Espacio de funciones: Puede ser un espacio vectorial. Por ejemplo, el conjunto de todas las funciones continuas en un intervalo dado forma un espacio vectorial. 

5. Espacio de vectores polares: Los vectores en un plano polar, definidos por un ángulo y una distancia del origen, forman un espacio vectorial. La suma de vectores polares se realiza combinando las distancias y sumando los ángulos, y la multiplicación por un escalar afecta solo a la distancia.

Hay muchos más tipos de espacios vectoriales. Cada espacio vectorial tiene sus propias propiedades y características distintivas, pero todos comparten las propiedades básicas que definen a un espacio vectorial.

***Subespacios vectoriales. Subespacios generadores***

Un subespacio vectorial es un conjunto de vectores que forma un espacio vectorial en sí mismo, es decir, cumple con todas las propiedades de un espacio vectorial. Un subespacio vectorial está contenido dentro de un espacio vectorial más grande.

Para que un conjunto de vectores sea un subespacio vectorial, debe cumplir las siguientes condiciones:

1. Contiene el vector cero: El conjunto debe incluir el vector cero, que actúa como el elemento neutro para la adición de vectores.

2. Cerrado bajo la adición: Si tomamos dos vectores del conjunto y los sumamos, el resultado también debe estar en el conjunto.

3. Cerrado bajo la multiplicación por escalar: Si tomamos un vector del conjunto y lo multiplicamos por un escalar, el resultado también debe estar en el conjunto.

Un ejemplo común de subespacio vectorial es el subespacio generado. El subespacio generado es el conjunto de todos los vectores que pueden ser expresados como combinaciones lineales de un conjunto dado de vectores llamados generadores. Los generadores son los vectores que "generan" el subespacio.

Para encontrar el subespacio generado por un conjunto de vectores, simplemente necesitamos encontrar todas las combinaciones lineales posibles de esos vectores. Por ejemplo, si tenemos los vectores v1 = (1, 0) y v2 = (0, 1), el subespacio generado estaría compuesto por todos los vectores de la forma (x, y) donde x e y son escalares.

El subespacio generado puede ser representado por el conjunto de todas las combinaciones lineales de los generadores. En el ejemplo anterior, el subespacio generado sería el conjunto de todas las combinaciones lineales de v1 y v2:

Span(v1, v2) = {(x, y) | x, y son escalares}

Los subespacios generados pueden ser unidimensionales (una línea a través del origen), bidimensionales (un plano que pasa por el origen) o tener dimensiones superiores según la cantidad de vectores generadores y sus combinaciones lineales posibles.

En resumen, los subespacios vectoriales, como el subespacio generado, son conjuntos de vectores que cumplen con las propiedades de un espacio vectorial y pueden ser utilizados para representar y analizar diferentes estructuras algebraicas en matemáticas y ciencias de la computación.

```python
import numpy as np

# Conjunto de vectores generadores
v1 = np.array([1, 2, 3])
v2 = np.array([-1, 0, 2])

# Combinaciones lineales
c1 = 2
c2 = -1

# Subespacio generado
subespacio_generado = c1 * v1 + c2 * v2

print("Subespacio generado:", subespacio_generado)
```

***Combinación lineal. Dependencia e independencia lineal***

La combinación lineal es una operación que involucra la multiplicación de vectores por escalares y su suma. En términos más generales, dados vectores v1, v2, ..., vn y escalares c1, c2, ..., cn, la combinación lineal se define como:

c1 * v1 + c2 * v2 + ... + cn * vn

La combinación lineal permite crear nuevos vectores a partir de una combinación ponderada de vectores existentes.

La dependencia e independencia lineal son conceptos relacionados con la combinación lineal y nos permiten determinar si un conjunto de vectores es linealmente dependiente o linealmente independiente.

Un conjunto de vectores se considera linealmente dependiente si existen escalares no todos nulos, c1, c2, ..., cn, tales que la combinación lineal de los vectores resulta en el vector cero (el vector cuyas componentes son todas cero). Esto se expresa como:

c1 * v1 + c2 * v2 + ... + cn * vn = 0

Por otro lado, un conjunto de vectores se considera linealmente independiente si la única combinación lineal que da como resultado el vector cero es aquella en la que todos los escalares son cero. Esto se expresa como:

c1 * v1 + c2 * v2 + ... + cn * vn = 0 implica c1 = c2 = ... = cn = 0

En resumen, la dependencia lineal implica que al menos uno de los vectores en el conjunto se puede expresar como una combinación lineal de los demás vectores, mientras que la independencia lineal implica que ninguno de los vectores en el conjunto se puede expresar como una combinación lineal de los demás vectores.


```python
import numpy as np

# Vectores de ejemplo
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 8, 9])

# Combinación lineal de los vectores
comb_lin = 2 * v1 + (-1) * v2 + 0.5 * v3

print("Combinación lineal: ", comb_lin)

# Verificar si los vectores son linealmente independientes
matrix = np.vstack((v1, v2, v3)).T
rank = np.linalg.matrix_rank(matrix)

if rank == len(matrix.T):
    print("Los vectores son linealmente independientes.")
else:
    print("Los vectores son linealmente dependientes.")
```

```python
import numpy as np

# Vectores de ejemplo
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])

# Verificar si los vectores son linealmente independientes
matrix = np.vstack((v1, v2, v3)).T
rank = np.linalg.matrix_rank(matrix)

if rank == len(matrix.T):
    print("Los vectores son linealmente independientes.")
else:
    print("Los vectores son linealmente dependientes.")
```

***Base y dimensión.***

La base y la dimensión son conceptos fundamentales asociados a los vectores y espacios vectoriales. Explicación breve de cada uno:

Base: Una base es un conjunto de vectores que son linealmente independientes y generan todo el espacio vectorial. En otras palabras, un conjunto de vectores forma una base si cada vector en el espacio puede expresarse de manera única como una combinación lineal de los vectores de la base. 

Dimension: La dimensión de un espacio vectorial es el número de vectores en cualquier base del espacio. Representa el "tamaño" o la cantidad de elementos necesarios para generar todos los vectores del espacio. La dimensión se denota como dim(V), donde V es el espacio vectorial.

Por ejemplo, considera el espacio vectorial R³ (espacio tridimensional). Una base común para este espacio está compuesta por los vectores canónicos: e₁ = [1, 0, 0], e₂ = [0, 1, 0] y e₃ = [0, 0, 1]. Estos vectores son linealmente independientes y cualquier vector en R³ puede expresarse como una combinación lineal de estos vectores. Por lo tanto, esta base tiene una dimensión de 3, ya que se necesitan 3 vectores para generar todo R³.

Es importante destacar que un espacio vectorial puede tener múltiples bases, pero todas las bases tienen el mismo número de vectores, lo que se conoce como la dimensión del espacio. Además, el concepto de dimensión se aplica tanto a espacios vectoriales finitos como a espacios vectoriales infinitos, como los espacios de funciones. 

```python
import numpy as np

# Definir una matriz que representa un conjunto de vectores
matriz = np.array([[1, 2, 0], [0, 1, -1], [-1, 0, 2]])

# Obtener la dimensión del espacio vectorial
dimension = np.linalg.matrix_rank(matriz)

# Obtener una base del espacio vectorial
base = np.linalg.qr(matriz)[0]

print("La dimensión del espacio vectorial es:", dimension)
print("Una base del espacio vectorial es:")
for vector in base.T:
    print(vector)
```

```python
import numpy as np

# Definir una lista de vectores
vectores = [
    np.array([1, 0, 0]),
    np.array([0, 1, 0]),
    np.array([1, 1, 1]),
    np.array([2, -1, 1])
]

# Construir una matriz con los vectores como columnas
matriz = np.column_stack(vectores)

# Obtener la dimensión del espacio vectorial
dimension = np.linalg.matrix_rank(matriz)

# Obtener una base del espacio vectorial
_, base = np.linalg.qr(matriz)

print("La dimensión del espacio vectorial es:", dimension)
print("Una base del espacio vectorial es:")
for vector in base.T:
    print(vector)
```

```python
import numpy as np

# Definir una lista de vectores
vectores = [
    np.array([1, 2, 0]),
    np.array([-1, 0, 1]),
    np.array([3, 1, 2]),
    np.array([2, 4, 1])
]

# Construir una matriz con los vectores como filas
matriz = np.vstack(vectores)

# Obtener la dimensión del espacio vectorial
dimension = np.linalg.matrix_rank(matriz)

# Obtener una base del espacio vectorial
_, base = np.linalg.qr(matriz.T)

print("La dimensión del espacio vectorial es:", dimension)
print("Una base del espacio vectorial es:")
for vector in base.T:
    print(vector)
```

***Coordenadas y matriz cambio de base***

Con vectores en un espacio vectorial, es útil poder representar esos vectores en diferentes bases. Las coordenadas y la matriz de cambio de base son herramientas utilizadas para realizar esta conversión.

Las coordenadas de un vector en una base son los coeficientes que se utilizan para expresar el vector como una combinación lineal de los vectores de la base. Por ejemplo, si tenemos una base B = {v1, v2, ..., vn} en un espacio vectorial V, y tenemos un vector v en V, las coordenadas de v con respecto a la base B se denotan como [v]_B y se definen como:

[v]_B = (a1, a2, ..., an)

Donde v = a1 * v1 + a2 * v2 + ... + an * vn.

La matriz de cambio de base es una matriz que nos permite convertir las coordenadas de un vector en una base a las coordenadas del mismo vector en otra base. Supongamos que tenemos dos bases B = {v1, v2, ..., vn} y B' = {u1, u2, ..., un} en el mismo espacio vectorial V. La matriz de cambio de base de B a B' se denota como P_{B',B} y se define como la matriz cuyas columnas son las coordenadas de los vectores de B' con respecto a la base B. Es decir:

P_{B',B} = [ [u1]_B , [u2]_B , ..., [un]_B ]

Una vez que tenemos la matriz de cambio de base, podemos convertir las coordenadas de un vector de B a B' utilizando la siguiente fórmula:

[v]_B' = P_{B',B} * [v]_B

Esto nos permite cambiar las coordenadas de un vector de una base a otra.

 Estas herramientas nos permiten comparar y transformar vectores en diferentes sistemas de referencia.

```python
import numpy as np

# Definición de las bases
B = np.array([[1, 0], [0, 1]])  # Base canónica en R^2
B_prime = np.array([[2, 1], [1, -1]])  # Otra base en R^2

# Vector de coordenadas en la base canónica
v = np.array([3, 4])

# Cálculo de las coordenadas de v en la base canónica
v_coordinates_B = v

# Cálculo de las coordenadas de v en la base B_prime
P_B_prime_B = np.linalg.inv(B) @ B_prime  # Matriz de cambio de base
v_coordinates_B_prime = P_B_prime_B @ v_coordinates_B

print("Coordenadas de v en la base canónica:", v_coordinates_B)
print("Coordenadas de v en la base B_prime:", v_coordinates_B_prime)
```

```python
import numpy as np

# Definición de las bases
B = np.array([[1, 1], [1, -1]])  # Base en R^2
B_prime = np.array([[1, 2], [3, 4]])  # Otra base en R^2

# Vector de coordenadas en la base B
v_coordinates_B = np.array([2, -1])

# Cálculo de la matriz de cambio de base
P_B_prime_B = np.linalg.inv(B) @ B_prime

# Cálculo de las coordenadas de v en la base B_prime
v_coordinates_B_prime = P_B_prime_B @ v_coordinates_B

print("Matriz de cambio de base (P_B_prime_B):\n", P_B_prime_B)
print("Coordenadas de v en la base B_prime:", v_coordinates_B_prime)
```

En este ejemplo, tenemos dos bases en el espacio vectorial R^2: la base B y otra base B_prime. Queremos calcular las coordenadas de un vector v en la base B_prime y obtener la matriz de cambio de base P_B_prime_B.

Definimos las bases B y B_prime utilizando matrices. Luego, establecemos el vector de coordenadas v_coordinates_B en la base B. En este caso, v_coordinates_B = [2, -1].

Calculamos la matriz de cambio de base P_B_prime_B utilizando la fórmula P_B_prime_B = inv(B) @ B_prime. Esta matriz nos permite transformar coordenadas de la base B a la base B_prime.

Finalmente, multiplicamos la matriz de cambio de base P_B_prime_B por las coordenadas de v en la base B para obtener las coordenadas de v en la base B_prime. Esto se realiza mediante la expresión v_coordinates_B_prime = P_B_prime_B @ v_coordinates_B.

***Bases ortogonales y ortonormales***

Son tipos especiales de bases en los espacios vectoriales. Se caracterizan por tener vectores que son ortogonales entre sí o que además de ser ortogonales, tienen norma igual a 1. 

1. Bases ortogonales:
   - En un espacio vectorial, una base se considera ortogonal si los vectores que la componen son ortogonales entre sí. Esto significa que el producto escalar entre cualquier par de vectores de la base es igual a cero.
   - Para una base ortogonal B = {v1, v2, ..., vn}, se cumple que vi · vj = 0 para todo i ≠ j, donde · representa el producto escalar.
   - Una base ortogonal puede facilitar ciertos cálculos y transformaciones, como la proyección ortogonal y la descomposición de un vector en componentes ortogonales.

2. Bases ortonormales:
   - Una base se considera ortonormal si además de ser ortogonal, los vectores tienen norma igual a 1. Esto significa que todos los vectores de la base son unitarios y ortogonales entre sí.
   - Para una base ortonormal B = {v1, v2, ..., vn}, se cumple que vi · vj = 0 para todo i ≠ j, y ||vi|| = 1 para todo i, donde · representa el producto escalar y ||v|| representa la norma del vector v.
   - Las bases ortonormales son especialmente útiles en aplicaciones como transformaciones lineales, resolución de sistemas de ecuaciones lineales y representación de señales en el análisis de Fourier.

Algunos ejemplos de bases ortogonales y ortonormales incluyen las bases canónicas en el espacio euclidiano R^n, las bases de los polinomios de Legendre y las bases de funciones sinusoidales en el análisis de Fourier.

Es importante destacar que no todos los espacios vectoriales tienen bases ortogonales o ortonormales. 

```python
import numpy as np

# Definir los vectores de la base ortogonal
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])

# Verificar la ortogonalidad
dot_product_12 = np.dot(v1, v2)
dot_product_13 = np.dot(v1, v3)
dot_product_23 = np.dot(v2, v3)

print("Producto escalar v1 · v2:", dot_product_12)
print("Producto escalar v1 · v3:", dot_product_13)
print("Producto escalar v2 · v3:", dot_product_23)
```
