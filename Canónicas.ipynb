# ***Las cónicas***

Son un conjunto de curvas obtenidas al cortar un cono de doble hoja con un plano. Incluyen el círculo, la elipse, la parábola y la hipérbola. 

El círculo es una cónica especial con todos los puntos equidistantes del centro. La elipse es una curva cerrada con dos ejes y simetría. La parábola es una curva abierta con un eje de simetría, y la hipérbola tiene dos ramas abiertas.

Las propiedades de una elipse incluyen las coordenadas del centro, las longitudes de los semiejes, la distancia focal, la distancia entre los focos, la longitud de la cuerda focal, la excentricidad y el área de la elipse. Estas propiedades se calculan a partir de los valores de "h", "k", "a" y "b", que representan el centro y las longitudes de los semiejes de la elipse.

¿Cómo se obtienen las secciones cónicas?

La superficie cónica se obtiene haciendo rotar una recta denominada generatriz alrededor de un punto fijo llamado vértice mantiendo otro punto constantemente sobre una circunferencia llamada directiz situaa en un plano perpendicular al eje y condicionada a que su centro esté sobre el eje.

Tipos de cónicas

El círculo es una cónica especial que se forma cuando el plano de corte es perpendicular al eje del cono. Es una curva cerrada en la que todos los puntos están equidistantes del centro. Su ecuación general es (x - h)² + (y - k)² = r², donde (h, k) representa el centro del círculo y r es su radio.

La elipse se produce cuando el plano de corte es oblicuo y corta ambas hojas del cono. Es una curva cerrada y simétrica con dos ejes: el eje mayor (2a) y el eje menor (2b). Su ecuación general es ((x - h)²/a²) + ((y - k)²/b²) = 1, donde (h, k) es el centro de la elipse y a y b son las longitudes de los semiejes.

La parábola se forma cuando el plano de corte es paralelo a una de las hojas del cono. Es una curva abierta y simétrica con un eje de simetría. Su ecuación general es y = a(x - h)² + k, donde (h, k) es el vértice de la parábola y a determina la apertura y dirección de la curva.

La hipérbola ocurre cuando el plano de corte atraviesa ambas hojas del cono. Es una curva abierta con dos ramas simétricas. Tiene dos ejes: el eje transverso (2a) y el eje conjugado (2b). Su ecuación general es ((x - h)²/a²) - ((y - k)²/b²) = 1, donde (h, k) es el centro de la hipérbola y a y b determinan la forma y la orientación de la curva.

¿Cómo se diferenciar los tipos de cónicas

Si el discriminmante es igual a cero, entonces la cónica es una circunferencia. Si el discriminante es mayor que cero, etonces la cónica es una elipse si A y C tienen el mismo signo, o una hipérbola si tienen signos opuestos. Si el discriminante es menor que cero, entonces la cónica es una parábola.


# ***La parábola***


Es una curva simétrica y curva que pertenece a la familia de las cónicas. Puede ser descrita como el conjunto de puntos equidistantes a un foco y a una recta llamada directriz en un plano.

La ecuación general de una parábola en un sistema de coordenadas cartesianas tiene la forma y = ax^2 + bx + c, donde a, b y c son constantes. Esta ecuación puede presentar distintas formas dependiendo de la posición y orientación de la parábola. Algunas de las formas más comunes son:

- Parábola vertical con vértice en el origen (0,0): y^2 = 4px, donde p es la distancia entre el foco y el vértice. La directriz es la línea x = -p.

- Parábola vertical con vértice en (h,k): (y-k)^2 = 4p(x-h), donde (h,k) es el vértice de la parábola. La directriz es la línea x = h - p.

- Parábola horizontal con vértice en el origen (0,0): x^2 = 4py, donde p es la distancia entre el foco y el vértice. La directriz es la línea y = -p.

- Parábola horizontal con vértice en (h,k): (x-h)^2 = 4p(y-k), donde (h,k) es el vértice de la parábola. La directriz es la línea y = k - p.

En una parábola, el vértice es el punto más alto o más bajo de la curva, dependiendo de si la parábola se abre hacia arriba o hacia abajo. El foco y la directriz determinan la forma y posición de la parábola. La distancia entre el foco y el vértice se conoce como distancia focal (FV), y la distancia entre el vértice y la directriz se denomina distancia focal (FD). Ambas distancias están relacionadas por la ecuación 4p = FV = FD.

Las parábolas encuentran aplicaciones en diversos campos como la física, la óptica, la ingeniería y las matemáticas aplicadas. Se utilizan en el diseño de reflectores parabólicos, en el análisis de trayectorias de proyectiles, en antenas parabólicas, en el estudio del movimiento de partículas y en muchas otras áreas.

```python
import math

def calcular_propiedades_parabola(h, k, a):
    # Coordenadas del vértice
    vertice = (h, k)
    print("Coordenadas del vértice:", vertice)
    
    # Distancia focal
    c = a / 4
    print("Distancia focal (c):", c)
    
    # Longitud de la cuerda focal
    longitud_cuerda_focal = abs(4 * a)
    print("Longitud de la cuerda focal:", longitud_cuerda_focal)
    
    # Eje de simetría
    eje_simetria = h
    print("Eje de simetría:", eje_simetria)

# Solicitar al usuario los valores de h, k y a
h = float(input("Ingrese el valor de h: "))
k = float(input("Ingrese el valor de k: "))
a = float(input("Ingrese el valor de a: "))

# Calcular y mostrar las propiedades de la parábola
calcular_propiedades_parabola(h, k, a)
```

# ***Hipérbola***

La hipérbola es una figura geométrica definida como el conjunto de puntos en el plano cuya diferencia de distancias a dos puntos fijos, llamados focos, es constante en valor absoluto.

La hipérbola tiene varios elementos característicos:

1. Focos: Son los puntos fijos F y F'.

2. Eje focal: Es la recta que pasa por los focos.

3. Eje secundario: Es la mediatriz del segmento FF'.

4. Centro: Es el punto de intersección de los ejes.

5. Vértices: Los puntos A y A' son los puntos de intersección de la hipérbola con el eje focal.

6. Radios vectores: Son los segmentos que van desde un punto de la hipérbola a los focos: PF y PF'.

7. Distancia focal: Es el segmento FF' de longitud 2c.

8. Eje mayor: Es el segmento AA' de longitud 2a.

9. Eje menor: Es el segmento BB' de longitud 2b.

Los puntos B y B' se obtienen como intersección del eje imaginario con una circunferencia que tiene uno de los vértices como centro y c como radio.

10. Ejes de simetría: Son las rectas que contienen al eje focal o al eje secundario.

11. Asintotas: Son las rectas con ecuaciones y = -b/(ax) y y = b/(ax).

12. Excentricidad: Es un parámetro que indica la abertura de la hipérbola. Siempre es mayor que 1 en el caso de las hipérbolas. La excentricidad se define como e = c/a, donde c es mayor o igual que a, y e es mayor o igual que 1.

La excentricidad es una medida de qué tan alejada está la hipérbola de ser una figura circular. Cuanto mayor sea el valor de la excentricidad, más abierta será la hipérbola.

```python
import math

def calcular_propiedades_hiperbola(h, k, a, b):
    # Coordenadas del centro
    centro = (h, k)
    print("Coordenadas del centro:", centro)
    
    # Longitudes de los semiejes
    semieje_a = a
    semieje_b = b
    print("Longitud del semieje transverso (a):", semieje_a)
    print("Longitud del semieje conjugado (b):", semieje_b)
    
    # Distancia focal
    c = math.sqrt(a**2 + b**2)
    print("Distancia focal (c):", c)
    
    # Distancia entre los focos
    distancia_focos = 2 * c
    print("Distancia entre los focos:", distancia_focos)
    
    # Excentricidad
    excentricidad = c / a
    print("Excentricidad (e):", excentricidad)
    
    # Eje transverso y eje conjugado
    eje_transverso = 2 * a
    eje_conjugado = 2 * b
    print("Longitud del eje transverso:", eje_transverso)
    print("Longitud del eje conjugado:", eje_conjugado)

# Solicitar al usuario los valores de h, k, a y b
h = float(input("Ingrese el valor de h: "))
k = float(input("Ingrese el valor de k: "))
a = float(input("Ingrese el valor de a: "))
b = float(input("Ingrese el valor de b: "))

# Calcular y mostrar las propiedades de la hipérbola
calcular_propiedades_hiperbola(h, k, a, b)
```

# ***Elipse***

Una elipse es una curva cerrada y simétrica que pertenece a la familia de las cónicas. Se define como el conjunto de puntos en un plano donde la suma de las distancias a dos puntos fijos, llamados focos, es constante.

La elipse tiene varios elementos característicos:

1. Focos: Son los puntos fijos F y F'.

2. Eje mayor: Es el segmento de longitud máxima que pasa por los focos y se extiende hasta los puntos de intersección de la elipse con su eje principal.

3. Eje menor: Es el segmento de longitud mínima que es perpendicular al eje mayor y se extiende hasta los puntos de intersección de la elipse con su eje secundario.

4. Vértices: Son los puntos de intersección de la elipse con su eje mayor.

5. Centro: Es el punto de intersección de los ejes mayor y menor. Es el punto donde se encuentran las simetrías de la elipse.

6. Semi-ejes: Son la mitad de las longitudes del eje mayor y el eje menor. Se denotan como a (semieje mayor) y b (semieje menor).

7. Excentricidad: Es un parámetro que indica qué tan alejada está la elipse de ser una figura circular. Se calcula como la distancia entre un foco y el centro dividida por el semieje mayor: e = c/a, donde c es la distancia entre un foco y el centro de la elipse.

8. Distancia focal: Es la distancia entre cada foco y el centro de la elipse. Se denota como c.

Las elipses tienen muchas aplicaciones en matemáticas, física y otras áreas científicas. Son utilizadas en la descripción de órbitas planetarias, trayectorias de satélites, diseño de lentes y reflectores, y en la modelización de fenómenos naturales y estructuras geométricas.

```python
import math

def calcular_propiedades_elipse(h, k, a, b):
    # Coordenadas del centro
    centro = (h, k)
    print("Coordenadas del centro:", centro)
    
    # Longitudes de los semiejes
    semieje_a = a
    semieje_b = b
    print("Longitud del semieje mayor (a):", semieje_a)
    print("Longitud del semieje menor (b):", semieje_b)
    
    # Distancia focal
    c = math.sqrt(a**2 - b**2)
    print("Distancia focal (c):", c)
    
    # Distancia entre los focos
    distancia_focos = 2 * c
    print("Distancia entre los focos:", distancia_focos)
    
    # Longitud de la cuerda focal
    longitud_cuerda_focal = 2 * a
    print("Longitud de la cuerda focal:", longitud_cuerda_focal)
    
    # Excentricidad
    excentricidad = c / a
    print("Excentricidad (e):", excentricidad)
    
    # Área de la elipse
    area_elipse = math.pi * a * b
    print("Área de la elipse:", area_elipse)

# Solicitar al usuario los valores de h, k, a y b
h = float(input("Ingrese el valor de h: "))
k = float(input("Ingrese el valor de k: "))
a = float(input("Ingrese el valor de a: "))
b = float(input("Ingrese el valor de b: "))

# Calcular y mostrar las propiedades de la elipse
calcular_propiedades_elipse(h, k, a, b)
```

# ***Circunferencia***

Una circunferencia es una figura geométrica que consiste en todos los puntos en un plano que están equidistantes de un punto fijo llamado centro. Es una curva cerrada y simétrica que no tiene ni principio ni fin.

La circunferencia tiene varias propiedades y elementos característicos:

1. Centro: Es el punto fijo que se encuentra en el medio de la circunferencia y equidistante de todos los puntos de la misma.

2. Radio: Es la distancia entre el centro de la circunferencia y cualquier punto de la misma. Todos los puntos de la circunferencia están a la misma distancia del centro.

3. Diámetro: Es el segmento que pasa por el centro de la circunferencia y tiene extremos en la circunferencia. El diámetro es igual a dos veces el radio.

4. Circunferencia unitaria: Es una circunferencia de radio igual a 1. Se utiliza como referencia en muchos cálculos y definiciones en matemáticas.

5. Longitud de la circunferencia: Es la medida del perímetro de la circunferencia. Se calcula utilizando la fórmula L = 2πr, donde r es el radio de la circunferencia y π (pi) es una constante aproximadamente igual a 3.14159.

6. Área del círculo: Es el espacio encerrado por la circunferencia. Se calcula utilizando la fórmula A = πr², donde r es el radio de la circunferencia y π es la constante pi.

Las circunferencias tienen numerosas aplicaciones en matemáticas, física, ingeniería y otras disciplinas. Son utilizadas en geometría analítica, trigonometría, cálculo, en el diseño de ruedas, engranajes, y en problemas que involucran simetría y propiedades geométricas.

```python
import math

def calcular_propiedades_circunferencia(h, k, r):
    # Coordenadas del centro
    centro = (h, k)
    print("Coordenadas del centro:", centro)
    
    # Radio
    radio = r
    print("Radio:", radio)
    
    # Diámetro
    diametro = 2 * radio
    print("Diámetro:", diametro)
    
    # Circunferencia
    circunferencia = 2 * math.pi * radio
    print("Circunferencia:", circunferencia)
    
    # Área
    area = math.pi * radio**2
    print("Área:", area)

# Solicitar al usuario los valores de h, k y r
h = float(input("Ingrese el valor de h: "))
k = float(input("Ingrese el valor de k: "))
r = float(input("Ingrese el valor de r: "))

# Calcular y mostrar las propiedades de la circunferencia
calcular_propiedades_circunferencia(h, k, r)
```
